```python
data['money'].plot(kind='hist', bins=30)
plt.title("Distribution of 'money'")
plt.xlabel("Money")
plt.ylabel("Frequency")
plt.show()
```

```python
data['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Proportion of Spam and Nonspam Emails")
plt.show()
```

```python
capitalAve_summary = data['capitalAve'].describe()
print(capitalAve_summary)
```

```python
mean_our = data['our'].mean()
median_our = data['our'].median()
std_our = data['our'].std()
print(f"Mean: {mean_our}, Median: {median_our}, Standard Deviation: {std_our}")
```

```python
min_hp = data['hp'].min()
max_hp = data['hp'].max()
print(f"Min: {min_hp}, Max: {max_hp}")
```

```python
percentiles = data['telnet'].quantile([0.25, 0.5, 0.75])
print(percentiles)
```

```python
mode_you = data['you'].mode()
print(f"Mode: {mode_you}")
```

```python
iqr_business = data['business'].quantile(0.75) - data['business'].quantile(0.25)
print(f"IQR: {iqr_business}")
```

```python
skew_credit = data['credit'].skew()
kurtosis_credit = data['credit'].kurtosis()
print(f"Skewness: {skew_credit}, Kurtosis: {kurtosis_credit}")
```

```python
correlation = data['capitalAve'].corr(data['capitalTotal'])
print(f"Correlation: {correlation}")
```

```python
range_num857 = data['num857'].max() - data['num857'].min()
print(f"Range: {range_num857}")
```

```python
variance_money = data['money'].var()
print(f"Variance: {variance_money}")
```
